```python
#  extract day, month, year from the date_published columns
p1["date_added"] = pd.to_datetime(p1['date_added'])
p1['day'], p1['month'], p1['year']  = p1['date_added'].dt.day, p1['date_added'].dt.month, p1['date_added'].dt.year
```

```python
def conv_to_int(df, col):
    df[col] =  df[col].fillna(0)
    df[col] = df[col].replace('',0, regex=True)
    np_of_df = df[col].values
    # convert to type as float
    np_of_df_float = np_of_df.astype(float)
    df[col] = np_of_df_float
    df[col] = df[col].astype(np.int64)
    return print("Completed!")
```

```python
list = ["day", "month", "year"]
for var in list:
    conv_to_int(p1, var)
```
